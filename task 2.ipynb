```python
# summary statistics
numeric_cols = ['Survived','Pclass','Sex_male','Embarked_Q','Embarked_S','Age_scaled','SibSp_scaled','Parch_scaled','Fare_scaled']

summary = df[numeric_cols].describe().T
summary['median'] = df[numeric_cols].median()
print(summary)
print('Computed summary statistics with median added')
```

```python
# This cell creates histograms and boxplots for numeric features


numeric_cols = ['Survived','Pclass','Sex_male','Embarked_Q','Embarked_S','Age_scaled','SibSp_scaled','Parch_scaled','Fare_scaled']

plt.figure(figsize=(12,8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3,3,i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(col)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3,3,i)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()

print('Plotted histograms and boxplots for numeric features')
```

```python
# Correlation matrix heatmap
numeric_cols = ['Survived','Pclass','Sex_male','Embarked_Q','Embarked_S','Age_scaled','SibSp_scaled','Parch_scaled','Fare_scaled']

corr = df[numeric_cols].corr()
print(corr)

plt.figure(figsize=(9,7))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Pairplot for a subset to keep it readable
subset_cols = [c for c in ['Survived','Pclass','Sex_male','Embarked_Q','Embarked_S','Age_scaled','SibSp_scaled','Parch_scaled','Fare_scaled'] if c in df.columns]
sns.pairplot(df[subset_cols], hue='Survived', corner=True, plot_kws=dict(s=15, alpha=0.6))
plt.show()
```

```python
# Identify Patterns, Trends, Anomalies, and Feature Inferences
# This cell computes pattern metrics and flags anomalies/outliers, plus simple feature-level inferences from visuals numerically.


# 1) Survival rates by key categorical-style features
surv_by_sex = df.groupby('Sex_male')['Survived'].mean()
surv_by_pclass = df.groupby('Pclass')['Survived'].mean()
surv_by_embark_Q = df.groupby('Embarked_Q')['Survived'].mean()
surv_by_embark_S = df.groupby('Embarked_S')['Survived'].mean()

# 2) Binned patterns (Age and Fare in quantiles)
df_tmp = df.copy()
df_tmp['Age_bin'] = pd.qcut(df_tmp['Age_scaled'], q=5, duplicates='drop')
df_tmp['Fare_bin'] = pd.qcut(df_tmp['Fare_scaled'], q=5, duplicates='drop')
surv_by_agebin = df_tmp.groupby('Age_bin')['Survived'].mean()
surv_by_farebin = df_tmp.groupby('Fare_bin')['Survived'].mean()

# 3) Family size and bins
family_size = df['SibSp_scaled'] + df['Parch_scaled']
# Use raw-ish indicator by thresholding scaled family size to approximate small vs large
small_family_mask = family_size <= family_size.quantile(0.75)
large_family_mask = ~small_family_mask
surv_small_family = df.loc[small_family_mask, 'Survived'].mean()
surv_large_family = df.loc[large_family_mask, 'Survived'].mean()

# 4) Simple anomaly flags via zscores on Fare_scaled
fare_z = df['Fare_scaled']
fare_outlier_idx = fare_z[fare_z.abs() > 3].index.tolist()

# 5) Interaction snapshot: Sex x Pclass survival
sex_class = df.groupby(['Sex_male','Pclass'])['Survived'].mean().unstack()

print(surv_by_sex)
print(surv_by_pclass)
print(surv_by_embark_Q)
print(surv_by_embark_S)
print(surv_by_agebin)
print(surv_by_farebin)
print(surv_small_family)
print(surv_large_family)
print(len(fare_outlier_idx))
print(sex_class)
print('Computed grouped survival rates, binned trends, family-size effects, outlier counts, and Sex x Pclass interaction')
```
